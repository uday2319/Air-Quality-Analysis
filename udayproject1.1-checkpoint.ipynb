```python
import pandas as pd

# Create sample data
df = {
    "Date": ["2025-08-01"] * 10,
    "Time": ["08:00"] * 10,
    "Location": [
        "New York", "Los Angeles", "Chicago", "Delhi", "London",
        "Sydney", "Beijing", "São Paulo", "Cairo", "Tokyo"
    ],
    "AQI": [85, 110, 72, 195, 55, 35, 210, 90, 130, 65],
    "pm2.5 ": [35.2, 50.1, 28.4, 95.3, 20.1, 10.5, 110.2, 40.6, 60.2, 22.5],
    "pm10": [60, 90, 55, 170, 40, 25, 200, 75, 110, 48],
    "NO₂ (ppb)": [22, 35, 18, 60, 12, 5, 70, 28, 45, 20],
    "SO₂ (ppb)": [6, 10, 4, 20, 3, 2, 25, 9, 15, 5],
    "CO (ppm)": [0.9, 1.2, 0.6, 1.5, 0.4, 0.3, 2.1, 1.0, 1.4, 0.7],
    "O₃ (ppb)": [30, 28, 25, 15, 33, 38, 10, 27, 22, 36],
    "Category": [
        "Moderate", "Unhealthy for Sensitive Groups", "Moderate", "Unhealthy", "Moderate",
        "Good", "Very Unhealthy", "Moderate", "Unhealthy for Sensitive Groups", "Moderate"]
}
df = pd.DataFrame(df)
df.to_csv("sample_air_quality_data.csv", index=['pm2.5','pm10','AQI'])

print("CSV file 'sample_air_quality_data.csv' has been created.")
```

```python
print(df.head())
```

```python
import pandas as pd
df['Date']=pd.to_datetime(df['Date'],errors='coerce')
df=df[df['Date'].dt.year== 2025]
df=df.dropna(subset=['AQI'])
print("rows after cleaning:",len(df))
```

```python
print("unique cities:",df['Location'].nunique())
```

```python
import pandas as pd
import numpy as np
top10=df.groupby('Location')['AQI'].mean().sort_values(ascending= True)
df.head(10)
```

```python
import matplotlib.pyplot as plt
top10.plot(kind='bar',figsize=(10,5))
plt.title("the top most polluted cities in 2025(by AQI)")
plt.ylabel("avearage AQI")
plt.xlabel("cities")
plt.show()
```

```python

Monthly_trend=df.groupby(df['Date'].dt.month)['AQI'].mean()
plt.bar(Monthly_trend.index,Monthly_trend.values,color='skyblue')
plt.title("global Monthly AQI trend in 2025")
plt.xlabel("Month")
plt.ylabel("Average AQI")
plt.xticks(range(1,13))
plt.grid(axis='y',linestyle='--')
plt.show()
```

```python
AQI_dist=df['Category'].value_counts()
AQI_dist.plot(kind='pie',autopct='%1.1f%%',figsize=(6,6))
plt.title("AQI category Distribution (2025)")
plt.show()
```
